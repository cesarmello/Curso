```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

food = pd.read_excel("fastfood.xlsx", sheet_name="fastfood")
food.head()
```

```python
food.groupby('restaurant').agg({'calories': 'mean', 'cal_fat': 'max'})
```

```python
plt.bar(food['restaurant'], food['cal_fat'])
plt.xticks(rotation=45)
plt.title("Media de Calorias por Restaurante")
plt.xlabel("Restaurante")
plt.ylabel("Media de Calorias")

plt.show()
```

```python
plt.hist(food['cal_fat'],bins=8)
plt.xticks(rotation=45)
plt.title("Media de Calorias por Restaurante")
plt.xlabel("Restaurante")
plt.ylabel("Media de Calorias")

plt.show()
```

```python
import seaborn as sns

# Plotar histograma e distribuição
sns.displot(food['cal_fat'], kde=True )
plt.show()
```

```python
sns.scatterplot(x='calories', y='sugar', size = 'protein',data=food, alpha=0.7)
plt.show()
```
